#### Split my data into a training and test set and create scatterplots, histograms, and a descriptive statistics table of the variables of interest.

```python
from sklearn.model_selection import train_test_split
```

```python
y = df_processed["salary"]
final = ["yearsworked","yearsrank"]
X_4 =df_processed[final]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_4, y, test_size=0.20, random_state=42)
```

```python
### STATSMODELS ###

X_train =  sm.add_constant(X_train)
X_test =  sm.add_constant(X_test)


# create a fitted model with all three features
model4 = sm.OLS(y_train, X_train).fit()

fpredictions = model4.predict(X_train)
```

```python
model1 = sm.load('model1_orig.pickle')
model3 = sm.load('model3_orig.pickle')
model4 = sm.load('model4_orig.pickle')
```

```python
std_residuals = statsmodels.regression.linear_model.RegressionResults.resid(model4)
```
